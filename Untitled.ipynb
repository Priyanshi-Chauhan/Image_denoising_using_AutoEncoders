# Data Preprocessing

```python
# Normalise the dataset
x_train=x_train/255
x_test=x_test/255
```

```python
x_train
```

```python
# Add some noise
noise_factor=0.3
noise_dataset=[]
for img in x_train:
    noisy_image = img + noise_factor * np.random.randn(*img.shape)         #(*img.shape)--> setting the dimensions
    noisy_image=np.clip(noisy_image,0,1)
    noise_dataset.append(noisy_image)
    
```

```python
noise_dataset=np.array(noise_dataset)
```

```python
plt.imshow(noise_dataset[10],cmap="gray")
```

```python
# Adding some noise to the test data
noise_factor=0.1
noise_test_dataset=[]
for img in x_test:
    noise_image = img + noise_factor* np.random.randn(*img.shape)
    noisy_image=np.clip(noisy_image,0,1)
    noise_test_dataset.append(noisy_image)
```

```python
noise_test_dataset=np.array(noise_test_dataset)
```

# Task -Build and train autoencoders deep learning model 

```python
autoencoders=tf.keras.models.Sequential()
```

```python
!pip install --upgrade keras
```

```python
autoencoders.add(tf.keras.layers.Conv2D(16,3,2,padding="same",input_shape=(28,28,1)))
```

```python
autoencoders.add(tf.keras.layers.Conv2D(8,3,2,padding="same"))

#encoded image
autoencoders.add(tf.keras.layers.Conv2D(8,3,1,padding="same"))

#decoder
autoencoders.add(tf.keras.layers.Conv2DTranspose(16,3,2,padding="same"))
autoencoders.add(tf.keras.layers.Conv2DTranspose(1,3,2,activation="sigmoid",padding="same"))
```

```python
autoencoders.compile(loss="binary_crossentropy",optimizer=tf.keras.optimizers.Adam(lr=0.001))
```

```python
autoencoders.summary()
```

### the output size matches with the size of input and the image size in the beginning itself has been compressed

# Evaluate Trained Model Performance

```python
autoencoders.fit(noise_dataset.reshape(-1,28,28,1),x_train.reshape(-1,28,28,1),epochs=10,batch_size=200,validation_data=(noise_test_dataset.reshape(-1,28,28,1),x_test.reshape(-1,28,28,1)))
```

# Evaluate Trained model performance

```python
evaluation=autoencoders.evaluate(noise_test_dataset.reshape(-1,28,28,1),x_test.reshape(-1,28,28,1))    # test loss is evaluated
```

```python
predicted=autoencoders.predict(noise_test_dataset[:10].reshape(-1,28,28,1))
```
